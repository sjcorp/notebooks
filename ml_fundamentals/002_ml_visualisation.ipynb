# Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a= np.arange(40,50)
b=np.arange(50,60)
```

```python
# Plt Scatter
plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Grpah in 2D')
```

### Scatter Plots

```python
# plt Scatter
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
```

### Sub Plots

```python
# Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

### Bar Plot

```python
# Bar Plot
x = [2,8,10]
y = [11, 16, 9]

x2 = [3,9,11]
y2 = [6, 15, 7]
plt.bar(x,y)
plt.bar(x2,y2, color = 'g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('x axis')
```

### Histogram

```python
# Histograms
sampl = np.random.uniform(low=1, high=30, size=(50,))
print(sampl)
plt.hist(sampl,bins=20)
plt.title("Histogram")
plt.show
```

### Box Plots

```python
plt.boxplot(sampl, vert = True, patch_artist=True)
```

### Pie Chart

```python
labels = 'Python','C++','Ruby','Java'
sizes = [215,213,130,234]
colors = ['gold','yellow','lightcoral','blue']
explode = (0.4,0,0,0) # Explode first slice

# Plot
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow=False)
plt.axis('equal')
plt.show
```

# Seaborn

Distribution Plots:
- distplot
- joinplot
- pairplot

Useful for Bivariate (2 Features) and Multivariate (More than 2 Features) Analysis

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

### Correlation with Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### JointPlot

#### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### Pair Plot (for multivariate analysis)

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

### Dist Plot

Helps to check the distributino of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=True,bins=10) # KDE Kernel Density Estimation (value vs. density)
```

## Categorical Plots (For Categorical Features)
- countplot
- boxplot
- violinplot
- countplot
- bar plot

### Count Plot

Simple graph of total counts

```python
sns.countplot("sex", data=df)
```

### Bar Plot

```python
sns.barplot(x='total_bill',y='sex', hue='day',data=df)
```

### Box Plot

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df, hue='smoker',palette='rainbow')
```

### Violin Plot

Allows us to see the distribution in terms of both - KDE and Box Plot

```python
sns.violinplot(x="total_bill",y="day",data=df,palette="rainbow")
```
